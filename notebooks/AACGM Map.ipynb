Produce a map of AACGM Latitude at ground level.

```python
import aacgmv2
import datetime as dt
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import cartopy.crs as ccrs 
```

```python
mpl.rcParams['font.size']      = 18
mpl.rcParams['font.weight']    = 'bold'
mpl.rcParams['axes.grid']      = True
mpl.rcParams['grid.linestyle'] = ':'
mpl.rcParams['figure.figsize'] = 1.5*np.array([8.0, 6.0])
```

```python
np.set_printoptions(formatter={'float_kind': lambda x:'{:.4f}'.format(x)})
```

```python
d_lat = 1.
d_lon = 1.

glats  = np.arange( -90, 90,d_lat)
glons  = np.arange(-180,180,d_lon)

glat_grid, glon_grid = np.meshgrid(glats,glons)

glat_grid_flat = glat_grid.flatten()
glon_grid_flat = glon_grid.flatten()

dtime = dt.datetime(2013, 11, 3)
alt   = 0.

shape = glat_grid.shape
```

```python
mlats = []
mlons = []
mlts  = []
for glat,glon in zip(glat_grid_flat,glon_grid_flat):
    result = aacgmv2.get_aacgm_coord(glat, glon, alt, dtime)
    #print(result)
    mlats.append(result[0])
    mlons.append(result[1])
    mlts.append( result[2])
```

```python
mlats = np.array(mlats)
mlons = np.array(mlons)
mlts  = np.array(mlts)

mlats.shape = shape
mlons.shape = shape
mlts.shape  = shape
```

```python
class nf(float):
    def __repr__(self):
        s = f'{self:.1f}'
        return f'{self:.0f}' if s[-1] == '0' else s
```

```python
figsize = np.array([10,8])*2
fig = plt.figure(figsize=figsize)
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
#ax = plt.axes(projection=ccrs.Robinson())

ax.set_global()
ax.coastlines()
ax.gridlines(draw_labels=True,color='black')

vmin, vmax = (-90., 90.)
cmap = 'jet'

levels = np.arange(-90,90,10)
CS = ax.contour(glon_grid, glat_grid, mlats, levels=levels,linewidths=3,
        vmin=vmin,vmax=vmax,cmap=cmap,
        transform=ccrs.PlateCarree())

# Recast levels to new class
#CS.levels = [nf(val) for val in CS.levels]

fmt = '%.0f'
ax.clabel(CS, CS.levels, inline=True, fmt=fmt, fontsize=15)
    
levels = np.arange(-90,90,1)
ax.contour(glon_grid, glat_grid, mlats, levels=levels,linewidths=0.5,
           vmin=vmin,vmax=vmax,cmap=cmap,
           transform=ccrs.PlateCarree())

date_str = dtime.strftime('%Y %b %d')
ax.text(0.5,1.08,'AACGM v2 Latitudes\n{!s} - {:.0f} km alt'.format(date_str,alt),
       ha='center',transform=ax.transAxes,fontdict={'size':'large'})

ax.text(-0.04, 0.55, 'Geographic Latitude', va='bottom', ha='center',
        rotation='vertical', rotation_mode='anchor',
        transform=ax.transAxes)

ax.text(0.5, -0.1, 'Geographic Longitude', va='bottom', ha='center',
        rotation='horizontal', rotation_mode='anchor',
        transform=ax.transAxes)

plt.show()

fig.savefig('aacgm.png',bbox_inches='tight')
```
